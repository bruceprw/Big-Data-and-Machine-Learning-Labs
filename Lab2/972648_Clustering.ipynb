# Iris dataset
The data set contains 50 samples from 3 species of Iris flower (150 samples total) and measures 4 features: 
1 - Petal length
2 - Petal width
3 - Sepal length
4 - Sepal width


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
```

# Task 2.1

```python
data = np.load('Iris_data.npy')
labels = np.load('Iris_labels.npy')
plt.figure()
plt.scatter(data[:,0], data[:,1], c=labels)

plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")

plt.title("Scatter plot of Iris data")
```

# Task 2.2
I will use k=3 as there are 3 species of iris

I use the fit_predict function of my KMeans clustering object to predict which cluster all of my samples belong to.

I then draw these on a scatter plot


```python
km = KMeans(n_clusters=3)

yk = km.fit_predict(data)

plt.figure()
plt.scatter(data[yk == 0, 0], data[yk == 0, 1])
plt.scatter(data[yk == 1, 0], data[yk == 1, 1])
plt.scatter(data[yk == 2, 0], data[yk == 2, 1])
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], marker='X')

plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")

plt.title("K-means clustering of Iris data")


```

# Task 2.4 New Dataset
I will use the elbow method to determine the correct number of clusters

```python
data = np.load('task2_4.npy')
K = range(1,10)
distortions = []
for k in K:
    kModel = KMeans(n_clusters=k)
    kModel.fit(data)
    distortions.append(kModel.inertia_)

plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

This graph shows that our optimal k is at 6

```python
kModel = KMeans(n_clusters=6)
kOut = kModel.fit_predict(data)
plt.ion()
plt.figure()
plt.scatter(data[kOut == 0, 0], data[kOut == 0, 1])
plt.scatter(data[kOut == 1, 0], data[kOut == 1, 1])
plt.scatter(data[kOut == 2, 0], data[kOut == 2, 1])
plt.scatter(data[kOut == 3, 0], data[kOut == 3, 1])
plt.scatter(data[kOut == 4, 0], data[kOut == 4, 1])
plt.scatter(data[kOut == 5, 0], data[kOut == 5, 1])



plt.scatter(kModel.cluster_centers_[:,0], kModel.cluster_centers_[:,1], marker='X')
```
